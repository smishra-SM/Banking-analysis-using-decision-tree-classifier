```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
my_data = pd.read_csv("UniversalBank.csv", delimiter=",")
my_data.head(10)
```

```python
X = my_data[['Income', 'CCAvg', 'Mortgage', 'Personal Loan']].values
y = my_data["CreditCard"]
```

```python
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
train = DecisionTreeClassifier(criterion="entropy", max_depth = 5).fit(X_trainset,y_trainset)
```

```python
y_prediction = train.predict(X_testset)
```

```python
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_testset, y_prediction))
```
